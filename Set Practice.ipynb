```python
set = {"apple", "banana", "cherry"}
print(set)
```

```python
# Check if banana is present/not present in a set
set = {"apple", "banana", "cherry"}

print("banana" in set)
print("banana" not in set)
```

```python
#Add items using the .add() function
set = {"apple", "banana", "cherry"}

set.add("orange")

print(set)
```

```python
# Add items from one set into another with the .update() method

thisset = {"apple", "banana", "cherry"}
tropical = {"pineapple", "mango", "papaya"}

thisset.update(tropical)
print(thisset)
```

```python
# Remove items with .remove() and .discard() methods

thisset.discard("banana")
thisset.remove("papaya")

print(thisset)
```

```python
# Union
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)
```

```python
# Can also use | to do the same thing
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1 | set2
print(set3)
```

```python
# intersection() method returns a new set with only the items that are present in both sets
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.intersection(set2)
print(set3)
```

```python
# Can you & to do the same thing
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 & set2
print(set3)
```

```python
# difference() will return a new set with items in the first set that are NOT present in the other
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)
print(set3)
```

```python
# symmetric_difference() keeps only the elements that are NOT present in both sets
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)
print(set3)
```

```python
# Can use ^ to do the same thing
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 ^ set2
print(set3)
```
