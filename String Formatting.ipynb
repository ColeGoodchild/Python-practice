```python
# F-strings allow you to format selected parts of a string
txt = f"The price is 30 dollars"
print(txt)
```

```python
# Add a placeholder like a variable
price = 59
txt = f"The price is {price} dollars"
print(txt)
```

```python
# Add a modifier to format the value
price = 59
txt = f"The price is {price:.2f} dollars" # in this case, a fixed point number with 2 decimals
print(txt)
```

```python
# Perform operations in f strings
txt = f"The price is {20*56} dollars"
print(txt)
```

```python
# Add taxes
price = 40
tax = 0.25
txt = f"The price is {price + (price * tax)} dollars"
print(txt)
```

```python
# if-else statements in f strings
price = 49
txt = f"It is very {'Expensive' if price>50 else 'Cheap'}"

print(txt)
```

```python
# Execute functions
fruit = "apples"
txt = f"I love {fruit.upper()}"
print(txt)
```

```python
# Use comma as a thousand separator:
price = 59000000
txt = f"The price is {price:,} dollars"
print(txt)
```
