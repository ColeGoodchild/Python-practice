```python
# The try block lets you test a block of code for errors.

# The except block lets you handle the error.

# The else block lets you execute code when there is no error.

# The finally block lets you execute code, regardless of the result of the try- and except blocks.
```

```python
try:
    print(x)
except:
    print("An exception occurred")

# Since try block raises an error, the except block will execute
# Without try, the program will crash
```

```python
# Can define as many exception block as needed
try:
    print(x)
except NameError:
    print("Variable x is not defined")
except:
    print("Something else went wrong")
```

```python
# Else executes if no errors were raised
try:
    print("Hello")
except:
    print("Something went wrong")
else:
    print("Nothing went wrong")
```

```python
# finally will be executed regardless if the try block raises an error or not
try:
    print(x)
except:
    print("Something went wrong")
finally:
    print("Sequence finished")
```
