```python
# __init__() method assigns values to object properties or to perform operations that are necessary when the object is being created

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Cole", 100)

print(p1.name)
print(p1.age)
```

```python
# We use it because without it, we would have to set properties manually for each object, like this:
class Person:
    pass

p1 = Person()
p1.name = "Cole"
p1.age = 100

print(p1.name)
print(p1.age)
```

```python
# Setting default values with the __init__() method
class Person:
    def __init__(self, name, age=20):
        self.name = name
        self.age = age

p1 = Person("Cole")
p2 = Person("Pete", 30)

print(p1.name, p1.age)
print(p2.name, p2.age)
```

```python
# It can use as many parameters as we need
class Pet:
    def __init__(self, name, breed, age, personality):
        self.name = name
        self.breed = breed
        self.age = age
        self.personality = personality

p1 = Pet("Daisy", "Golden Retriever", 10, "Friendly")

print(p1.name)
print(p1.breed)
print(p1.age)
print(p1.personality)
```

```python
# Using "self" parameter to access class properties
class Pet:
    def __init__(self, name, animal):
        self.name = name
        self.animal = animal

    def greet(self):
        print("Woof! My name is " + self.name)

p1 = Pet("Daisy", "Dog")
p1.greet()
```

```python
# Self links the method to the specific object:
class Pet:
    def __init__(self, name):
        self.name = name

    def printname(self):
        print(self.name)

p1 = Pet("Daisy")
p1.printname()
```

```python
# Accessing multiple properties using 'self'
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        print(f"{self.year} {self.brand} {self.model}")

car1 = Car("Toyota", "Corolla", 2008)
car1.display_info()
```
