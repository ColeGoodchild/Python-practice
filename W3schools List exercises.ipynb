```python
# emptying the list, leaving it with no elements
thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)
```

```python
# looping through a list
thislist = ["apple", "banana", "cherry"]
for i in thislist:
    print(i)
```

```python
# looping by index numbers
thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
    print(thislist[i])
```

```python
# looping with whilte loops
thislist = ["apple", "banana", "cherry"]
i = 0
while i < len(thislist):
    print(thislist[i])
    i = i + 1
```

```python
# List comprehension
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
    if "a" in x:
        newlist.append(x)

print(newlist)

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist2 = [x for x in fruits if "a" in x]
print(newlist2)
```

```python
# Setting values to upper case
newlist = [x.upper() for x in fruits]
print(newlist)
```

```python
# sorting lists alphabetically
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort() #ascending by default
print(thislist)
```

```python
# or numerically
thislist = [100, 50, 65, 82, 23]
thislist.sort()
thislist
```

```python
# descending order now
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort(reverse=True)
print(thislist)
```

```python
#  Customizing sort function
def myfunc(n):
    return abs(n - 50)
thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```

```python
# Case insensitive order
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
```

```python
# Reversing order of a list, regardless of alphabet
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

```python
# Copying a list
thislist = ["banana", "Orange", "Kiwi", "cherry"]
mylist = thislist.copy()
print(mylist)
```

```python
# copying using list() method
thislist = ["banana", "Orange", "Kiwi", "cherry"]
list1 = list(thislist)
print(list1)
```

```python
# Joining list
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
#adds it to the end
```

```python
# Joining list by appending
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

for x in list2:
    list1.append(x)

print(list1)

# Let's do the reverse now

list3 = ["a", "b", "c"]
list4 = [1, 2, 3]

for x in list3:
    list4.append(x)

print(list4)
```

```python
# Same thing with with extend() method
list5 = [45, 21, 76, 12]
list6 = ["Cole", "George", "Ben"]

list5.extend(list6)
print(list5)

list5.count(45) # Returns the number of elements with the specified value
```
