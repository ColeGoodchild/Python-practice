```python
# Creating a parent class that child classes can inherit properties from
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

# Use the Person class to create an object, and then execute the printname method:

x = Person("John", "Silver")
x.printname()
```

```python
# Make a child class
class Student(Person):
    pass

x = Student("Mike", "Johnson")
x.printname()
```

```python
# When we add the __init__() method to a child class, it will no longer inherit the parent's __init__() method
class Student(Person):
    def __init__(self, fname, lname):
        Person.__init__(self, fname, lname) # We can do this to keep the inheritance of the parent class's __init
```
