```python
# Properties store data for each object created from the class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Cole", 30)

print(p1.name)
print(p1.age)
```

```python
# Modifying properties using variables
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Cole", 30)
print(p1.age)

p1.age = 26
print(p1.age)
```

```python
# Deleting properties with the 'del' keyword
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

p1 = Car("Chevy", "Cruze")
print(p1.make)

del p1.model
print(p1)
```

```python
# Properties defined outside methods belong to the class and are shared by all objects
class Person:
    species = "Human" #class property

    def __init__(self, name):
        self.name = name #instance property

p1 = Person("Cole")
p2 = Person("Pete")

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```

```python
# Methods can accept parameters just like regular functions
class Calculator:
    def add(self, a, b):
        return a + b

    def multiply(self, a, b):
        return a * b

calc = Calculator()
print(calc.add(5, 3))
print(calc.multiply(4,7))
```

```python
# Methods can modify the properties of an object
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def birthday(self):
        self.age += 1
        print(f"Happy Birthday! You are now {self.age}!")

p1 = Person("Cole", 30)
p1.birthday()
p1.birthday()
```

```python
# __str__() is a method that controls what is returned when the object is printed
class House:
    def __init__(self, style, YearBuilt):
        self.style = style
        self.YearBuilt = YearBuilt

    def __str__(self):
        return f"{self.style} ({self.YearBuilt})"

h1 = House("Colonial", 1991)
print(h1)
```

```python
# A class can have multiple methods that work together
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)
        print(f"Added: {song}")

    def remove_song(self, song):
        if song in self.songs:
            self.songs.remove(song)
            print(f"Removed: {song}")

    def show_songs(self):
        print(f"Playlist '{self.name}':")
        for song in self.songs:
            print(f"- {song}")

my_playlist = Playlist("Favorites")
my_playlist.add_song("Bohemian Rhapsody")
my_playlist.add_song("Layla")
my_playlist.show_songs()
```
