```python
dict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(dict)
```

```python
# Access dicts using the key name
print(dict["brand"])
```

```python
# Use len to see how many items a dictionary has
print(len(dict))
```

```python
print(type(dict))
```

```python
# Can also access with get() method
x = dict.get("brand")
print(x)
```

```python
# keys() will return all the keys
x = dict.keys()
print(x)
```

```python
# Adding an item to a dict
car = {
    "brand": "Ford",
    "model": "Sweden",
    "year": 1964
}

x = car.keys()
print(x) #before change

car["color"] = "white"
print(x) # after change
```

```python
# values() returns all the values in the dict
y = dict.values()
print(y)
```

```python
# Return key value pairs with items()
x = dict.items()
print(x)
```

```python
# change values by referring to its key name
car = {
    "brand": "Ford",
    "model": "Sweden",
    "year": 1964
}
car["brand"] = "chevy"
print(car)
```

```python
# update both with update()
car = {
    "brand": "Ford",
    "model": "Sweden",
    "year": 1964
}
car.update({"year": 2026})
print(car)
```

```python
# Remove item with pop()
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

```python
# Access all values in a dictionary
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
for x in thisdict:
    print(thisdict[x])
```

```python
# Loop through keys and values with items() method
for x, y in thisdict.items():
    print(x, y)
```

```python
# Nest dicts
family = {
    "child1": {
    "name" : "Sandy",
    "year": 2004
              },
    "child2" : {
        "name": "Peter",
        "year": 2008
    },
    "child3" : {
        "name": "John",
        "year": 2006
    }
}
print(family)
```

```python
# Looping through nested dicts
for x, obj in family.items():
    print(x)

    for y in obj:
        print(y + ':', obj[y])
```

```python
# Use "_" as the last case value if you want a block to execute when there are no other matches
day = 7
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday!!")
    case _:
        print("No work today, let's party!")
```

```python
# Use '|' to check for more than one value match in one case
day = 4
match day:
    case 1 | 2 | 3 | 4 | 5:
        print("work today ;(")
    case 6 | 7:
        print("I'm free!")
```
