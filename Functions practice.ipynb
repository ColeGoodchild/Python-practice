```python
# Arguments before "/" are positional only, and arguments after * are keyword only:
def someFunc(a, b, /, *, c, d):
    return a + b + c + d

result = someFunc(5, 10, c = 15, d = 20)
print(result)
```

```python
# *args and **kwargs allow functions to accept an unknown number of arguments
# If we don't know how many args our function will take, add a "*" before the parameter name

def family(*kids):
    print("The youngest child is " + kids[2])

family("Emil", "Jake", "Nick")
```

```python
# *args lets our function accept any number of positional arguments

def example(*args):
    print("Type:", type(args))
    print("1st argument:", args[0])
    print("2nd argument:", args[1])
    print("All arguments:", args)

example("Bill", "Bob", "Joe")
```

```python
# Combining regular params with *args
def greetings(greeting, *names):
    for name in names:
        print(greeting, name)

greetings("Hello", "John", "Joe", "Alice")
```

```python
# use *args to create flexible functions
def sum(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum(1,2,3))
print(sum(10,20,30,40))
print(sum(5))
```

```python
# Finding the maximum value
def max(*numbers):
    if len(numbers) == 0:
        return None
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

print(max(3, 7, 2, 9, 1))
```

```python
# flexible keyword arguments function example
def family(**kid):
    print("His last name is " + kid["lname"])

family(fname = "Tobias", lname = "Lakin")
```

```python
# **kwargs allows a function to accept any number of keyword arguments

def my_function(**myvar):
  print("Type:", type(myvar))
  print("Name:", myvar["name"])
  print("Age:", myvar["age"])
  print("All data:", myvar)

my_function(name = "Tobias", age = 30, city = "Bergen")
```

```python
# We can combine *args and **kwargs BUT the order must be 1. regular params 2. *args 3. **kwargs
def my_function(title, *args, **kwargs):
  print("Title:", title)
  print("Positional arguments:", args)
  print("Keyword arguments:", kwargs)

my_function("User Info", "Emil", "Tobias", age = 25, city = "Oslo")
```

```python
x = "global"

def outer():
  x = "enclosing"
  def inner():
    x = "local"
    print("Inner:", x)
  inner()
  print("Outer:", x)

outer()
print("Global:", x)
```

```python
# Decorators takes another function as input and returns a new function
def changecase(func):
    def myinner():
        return func().upper()
    return myinner

@changecase
def myfunc():
    return "Hello Mark"

print(myfunc())
```

```python
# One decorator for upper case, and one for adding a greeting
def changecase(func):
    def myinner():
        return func().upper()
    return myinner

def addgreeting(func):
    def myinner():
        return "Hello " + func() + " Have a good day!"
    return myinner

@changecase
@addgreeting
def myfunction():
    return "Cole"

print(myfunction())
```
